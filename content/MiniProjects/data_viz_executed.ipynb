```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
data_path = '../data/interim/combined_data.csv'
df = pd.read_csv(data_path)

df['countyfips'] = df['countyfips'].astype(str)
df['countyfips'] = '0' + df['countyfips']
df.head()
```

```python
fig_regplot = px.scatter(df, x="Median Household Income", y="BINGE_AdjPrev", trendline="ols")
fig_regplot.show()
```

### Median Household Income vs. Binge Drinking

This scatterplot compares county median household income with adjusted binge drinking prevalence. The trendline suggests a weak negative relationship: counties with higher median household income tend to have slightly lower binge drinking prevalence.

However, the relationship is not very strong, so income alone does not explain binge drinking rates. Other factors such as age, education, urban/rural differences, or local culture may also matter.
